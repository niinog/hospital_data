```python
import json
import pandas as pd
rows = []

for file in Path(RAW_FHIR_DIR).glob("*.json"):
    with open(file, "r") as f:
        bundle = json.load(f)

    # Each file is usually a FHIR Bundle with an "entry" list
    for entry in bundle.get("entry", []):
        resource = entry.get("resource", {})
        if resource.get("resourceType") == "Observation":
            rows.append(resource)



# Flatten nested JSON into columns
df = pd.json_normalize(rows)
```

```python
df.head()
```

```python
cols = [
    "id",
    "subject.reference",
    "encounter.reference",
    "effectiveDateTime",
    "code.coding",
    "code.text",
    "valueQuantity.value",
    "valueQuantity.unit",
]

obs_small = df[cols].copy()
obs_small.head()
```

```python
# Remove "urn:uuid:" prefix to match CSV IDs
obs_small["patient_id"] = obs_small["subject.reference"].str.replace(
    "urn:uuid:", "", regex=False
)

obs_small["encounter_id"] = obs_small["encounter.reference"].str.replace(
    "urn:uuid:", "", regex=False
)

# Convert datetime

obs_small["observation_datetime"] = pd.to_datetime(
    obs_small["effectiveDateTime"], errors="coerce"
)

obs_small[[
    "id",
    "patient_id",
    "encounter_id",
    "observation_datetime"
]].head()
```

```python
def extract_coding_fields(codings):
    """
    Take the value of code.coding (usually a list of dicts)
    and return (system, code, display). If missing, returns (None, None, None).
    """
    if isinstance(codings, list) and len(codings) > 0 and isinstance(codings[0], dict):
        c = codings[0]
        return c.get("system"), c.get("code"), c.get("display")
    return (None, None, None)

coding_df = obs_small["code.coding"].apply(extract_coding_fields).apply(pd.Series)
coding_df.columns = ["code_system", "code", "code_display"]

obs_small = pd.concat([obs_small, coding_df], axis=1)

obs_small[["code_system", "code", "code_display"]].head()
```

```python
obs_clean = obs_small[[
    "id",
    "patient_id",
    "encounter_id",
    "observation_datetime",
    "code_system",
    "code",
    "code_display",
    "valueQuantity.value",
    "valueQuantity.unit",
]].rename(columns={
    "id": "observation_id",
    "valueQuantity.value": "result_value",
    "valueQuantity.unit": "result_unit",
})

obs_clean.head()
```

```python
from src.ingest import load_patients_csv, load_observations_fhir
from src.transform import (
    clean_patients,
    clean_observations_fhir,
    join_patients_observations,
)


patients_raw = load_patients_csv()
patients = clean_patients(patients_raw)

obs_raw = load_observations_fhir()
obs_clean = clean_observations_fhir(obs_raw)

# Join
patient_obs = join_patients_observations(patients, obs_clean)

print("Joined shape:", patient_obs.shape)
patient_obs.head(10)
```
